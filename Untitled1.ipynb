```python
class InventoryManager:
    def __init__(self, dataframe):
        self.df = dataframe

    def daily_sales(self):
        return self.df.groupby('date')['quantity'].sum()

    def product_sales(self, product_name):
        return self.df[self.df['product'] == product_name]
```

```python
class DemandAnalyzer:
    def __init__(self, daily_sales):
        self.daily_sales = daily_sales

    def moving_average_forecast(self, days=7, safety_factor=1.2):
        recent = self.daily_sales.tail(days)
        return np.mean(recent) * safety_factor
```

```python
current_stock = 100

inventory_manager = InventoryManager(df)
daily_sales_data = inventory_manager.daily_sales()
demand_analyzer = DemandAnalyzer(daily_sales_data)
forecast = demand_analyzer.moving_average_forecast()

if current_stock < forecast:
    print(f"⚠️ UYARI: Tahmini talep {forecast:.0f}. Stok artırılmalı.")
else:
    print("✅ Stok yeterli.")
```

```python
plt.figure()
daily_sales_data.plot()
plt.title("Günlük Satış Trendi")
plt.xlabel("Tarih")
plt.ylabel("Satış Miktarı")
plt.show()
```
